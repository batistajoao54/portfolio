```python
# codigo para criar esse appweb

import streamlit as st 
import pandas as pd
import plotly.figure_factory as ff

df = pd.read_csv('dados.csv')

st.set_page_config(layout='wide')
pagina = st.sidebar.radio('Pagina', ['INÍCIO','ANÁLISE','DELETAR'])



#=======Pagina inicial=======================
if pagina == 'INÍCIO':
    st.image('top.png')
    
              
    c1, c2, c3, c4 = st.columns([1,2,2,1])
    with c1:
        data = st.date_input('DATA')
    with c2:
        nome = st.text_input('NOME DO CLIENTE')
    with c3:
        servico = st.text_input('SERVIÇO PRESTADO')
    with c4:
        valor = st.text_input('VALOR R$')
     
    lista_data = [data]
    lista_nome = [nome.upper()]
    lista_servico = [servico.upper()]
    lista_valor = [valor]
    

    tabela = pd.DataFrame()
    
    
    c1, c2, c3, c4, c5 = st.columns([7,1,1,1,1])
    with c1:
    
        add = st.button('SALVAR')
        tabela['DATA'] = lista_data
        tabela['CLIENTE'] = lista_nome
        tabela['SERVIÇO'] = lista_servico
        tabela['VALOR R$'] = lista_valor
        
        tabela['VALOR R$'] = tabela['VALOR R$'].str.replace(',', '.')
        tabela['VALOR R$'] = pd.to_numeric(tabela['VALOR R$'])
        
            
        if add:
            df = pd.concat([df,tabela])
            df.to_csv('dados.csv', index=False)
    
    st.title('')
    
    c1,c3,c4,c5,c6 = st.columns([1,1,1,1,1])
    with c1:
        st.write('')
        st.header('MÊS')
    with c3:
        mes = st.selectbox('Escolha um mês', ['02', '01', '03', '04', '05',
                                            '06', '07', '08', '09', '10', '11', '12'])          
    with c5:
        st.write('')
        st.header('ANO')
    with c6:
        ano = st.selectbox("Escolha um ano",['22'])
    
    
    #============================================================================================================
    df_copy = df.copy()
    df_copy['DIA'] = pd.to_datetime(df_copy['DATA']).dt.strftime('%d')
    df_copy['MES'] = pd.to_datetime(df_copy['DATA']).dt.strftime('%m')
    df_copy['ANO'] = pd.to_datetime(df_copy['DATA']).dt.strftime('%y')
    
    df_modificado = pd.DataFrame()
    
    df_select = df_copy[df_copy['MES'] == mes]
    df_select2 = df_select[df_select['ANO'] == ano]
    
    df_modificado['DATA'] = df_select2['DATA']
    df_modificado['CLIENTE'] = df_select2['CLIENTE']
    df_modificado['SERVIÇO'] = df_select2['SERVIÇO']
    df_modificado['VALOR R$'] = df_select2['VALOR R$']
    
    #============================================================================================================
    
    st.write('')
    c1,c3 = st.columns([3,5])
    total = len(df_modificado)
    with c1:
        st.metric('SERVIÇO REALIZADO', value= f'{total}')
         
        
    
    
    with c3:
        st.text("DADOS ADICIONADOS")
        fig = ff.create_table(df_modificado, height_constant=10)
        st.plotly_chart(fig, use_container_width=True)
    
                






#========Pagina de adicionar os dados=========
elif pagina == 'ANÁLISE':
    c1, c2, c3, c4, c5 = st.columns([1,1,20,1,1])
    with c3:
        st.image('manuntecao.png')
    





   
#=========Pagina de editar os dados=======
elif pagina == 'DELETAR':
    A1,A2,A3 = st.columns([1,3,1])
    with A2:
        st.title('**APAGANDO OS DADOS**')
    
    st.header('MÊS')
    mes = st.selectbox('Escolha um mês', ['janeiro', 'fevereiro', 'março', 'abril', 'maio',
                                        'junho', 'julho', 'agosto', 'setembro', 'novembro', 'dezembro'])
    
    st.dataframe(df,width=1000, height=400)
    
    linha = st.text_input('INDICE DA LINHA QUE DESEJA EXLUIR')
    if linha != '':
        indice = int(linha)
    
    
        b1,b2,b3 = st.columns([3,1,3])
        with b2:
            add = st.button('EXCLUIR')
        
        if add:
            df.drop(indice, inplace = True)
            df.to_csv('dados.csv', index=False)
            st.markdown('**Mude de aba para atualizar a tabela**')

    
    if st.checkbox('Download dos dados'):
        senha = st.text_input('digite a senha', type='password')
        if senha == '102132':
            @st.cache
            def convert_df(df):
            # IMPORTANT: Cache the conversion to prevent computation on every rerun
                return df.to_csv().encode('utf-8')

            csv = convert_df(df)

            st.download_button(
            label="Download CSV",
                data=csv,
                file_name='dados.csv',
                mime='text/csv',
            )
        
        elif senha == '':
            lh = 'los hermanos'
        
        else:
            st.header('**SENHA INCORRETA!, TENTE NOVAMENTE.')


```
