```python
import pandas as pd
import requests
```

```python
df = pd.read_excel('NBA_Atual.xlsx')
```

```python
dia = "11"
mes = '01'
ano = "2022"

url = f'https://www.espn.com.br/nba/calendario/_/data/{ano}{mes}{dia}'
res = requests.get(url)
res
```

```python
tabela = pd.read_html(res.text)
dados = tabela[0].copy()
```

```python
dados['DATA'] = f'{dia}-{mes}-{ano}'
modelo = dados[['DATA', 'matchup', 'Unnamed: 1', 'resultado']].copy()
modelo.columns = ['DATA', 'VISITANTE', 'CASA', 'RESULTADO']
modelo
```

```python
#separando o time visitante
visitante = list(modelo['VISITANTE'].copy())
visitante_ = []

for i in range(len(visitante)):
    time = visitante[i][-3:]
    visitante_.append(time)


#separando o time da casa
casa = list(modelo['CASA'].copy())
casa_ = []

for i in range(len(casa)):
    time = casa[i][-3:]
    casa_.append(time)


#separando o resultado do jogo
resultado = list(modelo['RESULTADO'].copy())
resultado_win = []
resultado_loser = []
ponto_win = []
ponto_loser = []


#separando a coluna resutando e tratando os dados
for i in range(len(resultado)):
    win_ponto = int(resultado[i].split(',')[0].strip()[-3:])
    loser_ponto = int(resultado[i].split(',')[1].strip()[-3:])
    
    win_resultado = resultado[i].split(',')[0].strip()[:3]
    loser_resultado = resultado[i].split(',')[1].strip()[:3]
    
    ponto_win.append(win_ponto)
    ponto_loser.append(loser_ponto)
    resultado_win.append(win_resultado)
    resultado_loser.append(loser_resultado)

#pegando a data dos jogos
data = list(modelo["DATA"]).copy()

#criando um novo DF e colocando as lista para formar nosso banco de dados
base_dados = pd.DataFrame()

base_dados['DATA'] = data
base_dados['VISITANTE'] = visitante_
base_dados['CASA'] = casa_
base_dados['VENCEDOR'] = resultado_win
base_dados['WIN_PONTOS'] = ponto_win
base_dados['LOSER_PONTOS'] = ponto_loser
base_dados['PERDEDOR'] = resultado_loser
```

```python
base_dados
```

```python
#atualizando os dados com concat
novo_dados = pd.concat([df,base_dados])
```
